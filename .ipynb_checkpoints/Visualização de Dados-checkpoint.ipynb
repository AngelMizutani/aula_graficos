## Gráfico de barras
* gráfico de colunas
* frequencia de valores
* dados categóricos

```python
# agrupar os dados das profissões
df_occupation = df.groupby(['Occupation']).size()
df_occupation
```

```python
# gráfico de barras
df_occupation.plot.bar()
```

```python
# comparar as profissões com o nível de ansiedade
group_occupation_anxiety = df.groupby(
    ['Occupation'])['Anxiety Level (1-10)'].mean()
group_occupation_anxiety.plot.bar()
```

```python
# profissões X consumo de cafeína
group_occupation_caffeine = df.groupby(
    ['Occupation'])['Caffeine Intake (mg/day)'].sum()
group_occupation_caffeine.plot.bar()
```

## Gráfico de setores
* pizza ou diagrama circular
* precisão não é necessária
* poucos atributos
* variáveis categóricas

```python
group_occupation_anxiety.plot.pie(autopct = '%1.1f%%')
```

```python
# gráfico por genero
group_gender = df.groupby(['Gender']).size()
group_gender.plot.pie()
```

## Histograma
* dados numéricos
* dados agrupados por intervalos de valores (bins)
* distribuição de frequencia

```python
df.head()
```

```python
# histograma das horas de sono
# kde = linha de densidade
# densidade: melhora visualização
srn.histplot(df['Sleep Hours'], 
            kde = True,
            bins = 15,
            color = 'purple').set_title('Horas de sono')
```

```python
# só a linha de densidade
srn.kdeplot(df['Sleep Hours'])
```

```python
# histograma da atividade física (Physical Activity (hrs/week))
srn.histplot(df['Physical Activity (hrs/week)'],
            bins = 10).set_title('Atividade física')
```

```python
# histograma comparando dois atributos
# engenheiros X horas de sono
srn.histplot(
    df.loc[df['Occupation'] == 'Engineer']['Sleep Hours']
).set_title('Horas de Sono dos Engenheiros')
            
```

```python
# Horas de sono dos artistas
srn.histplot(
    df.loc[df['Occupation'] == 'Artist']['Sleep Hours']
).set_title('Horas de Sono dos Artistas')
```

## Gráfico de dispersão
* scatter plot
* visualizar como duas variáveis se relacionam

```python
# utilizar outra base de dados
df_trees = pd.read_csv('trees.csv')
df_trees.shape
```

```python
df_trees.head()
```

```python
df_trees.tail()
```

```python
# comparar circunferencia X volume
plt.scatter(df_trees['Girth'], df_trees['Volume'])
plt.title('Comparação da circunferência pelo volume das árvores')

#rótulos
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

## Gráfico de linhas

## Visualização de vários gráficos em uma figura

```python
df.head()
```

```python
df_occupation
```

```python
# histograma do nivel de ansiedade de médicos
srn.histplot(
    df.loc[df['Occupation'] == 'Doctor']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em médicos')
```

```python
srn.histplot(
    df.loc[df['Occupation'] == 'Nurse']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em enfermeiros')
```

```python
srn.histplot(
    df.loc[df['Occupation'] == 'Teacher']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em professores')
```

```python
srn.histplot(
    df.loc[df['Occupation'] == 'Artist']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em artistas')
```

```python
#imprimir todos os gráficos em uma figura
# Criar uma figura para hospedar os gráficos
plt.figure()

#plt.subplot(num_linhas, num_colunas, posição)
plt.subplot(2, 2, 1)
srn.histplot(
    df.loc[df['Occupation'] == 'Doctor']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em médicos')

plt.subplot(2, 2, 2)
srn.histplot(
    df.loc[df['Occupation'] == 'Nurse']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em enfermeiros')

plt.subplot(2, 2, 3)
srn.histplot(
    df.loc[df['Occupation'] == 'Teacher']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em professores')

plt.subplot(2, 2, 4)
srn.histplot(
    df.loc[df['Occupation'] == 'Artist']['Anxiety Level (1-10)'],
    bins = 9
).set_title('Nível de ansiedade em artistas')

#ajustar o layout para evitar sobreposição
plt.tight_layout()
```
